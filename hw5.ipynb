### Excercise B. 6 $\mathit{Use \ the \ Trapezoidal \ method \ to \ approximate \ the \ integral.}$ (integrate_exp.py)

Plot $e^{-x^2}$ for $x$ from $-10$ to $10$:

```python
p3.plot(100)
```

The funtion $e^{-x^2}$ is symmetric about the $y-axis$ so changing the bounds of the integral $\int_{-\infty}^\infty e^{-x^2}dx$ to $\int_0^\infty e^{-x^2}dx$ gives half the result. To make the integrals equal eachother, multiply the second by $2$: $$\int_{-\infty}^\infty e^{-x^2}dx = 2\int_0^\infty e^{-x^2}dx$$

The table shows the approximations of the integral for various $L$s and $n$s.

```python
p3.table()
```

The next table shows the error of these approximations from the known analytic solution to the integral, which is $\sqrt{\pi}$.
